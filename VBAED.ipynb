```python

df_encoded['Date'] = pd.to_datetime(df_encoded['Date'], dayfirst=True, errors='coerce')
df_encoded = df_encoded.sort_values(by=['Area', 'Date']).reset_index(drop=True)
unique_areas = df_encoded['Area'].unique()
np.random.seed(42) 
np.random.shuffle(unique_areas)
train_size = int(len(unique_areas) * 0.7)
val_size   = int(len(unique_areas) * 0.15)

train_areas = unique_areas[:train_size]
val_areas   = unique_areas[train_size:train_size + val_size]
test_areas  = unique_areas[train_size + val_size:]

df_train_final = df_encoded[df_encoded['Area'].isin(train_areas)].copy()
df_val_final   = df_encoded[df_encoded['Area'].isin(val_areas)].copy()
df_test_final  = df_encoded[df_encoded['Area'].isin(test_areas)].copy()

print(f"Training: {len(df_train_final)} records  ({len(train_areas)} areas)")
print(f" Validation: {len(df_val_final)} records  ({len(val_areas)} areas)")
print(f" Testing: {len(df_test_final)} records  ({len(test_areas)} areas)")

 
def extract_X_y(df):
    X_df = df.drop(columns=["CCME_WQI", "Country", "Area", "Date", "Year", "CCME_Values"])
    y_df = df["CCME_WQI"].astype(cat_type).cat.codes.values
    return X_df, y_df

X_train, y_train = extract_X_y(df_train_final)
X_val,   y_val   = extract_X_y(df_val_final)
X_test,  y_test  = extract_X_y(df_test_final)

 
X_train.columns = [col.replace(' ', '_').replace('/', '_').replace('(', '').replace(')', '') for col in X_train.columns]
X_val.columns   = X_train.columns
X_test.columns  = X_train.columns
 
num_features = ["Ammonia_mg_l", "Biochemical_Oxygen_Demand_mg_l", "Dissolved_Oxygen_mg_l", 
                "Orthophosphate_mg_l", "pH_ph_units", "Temperature_cel", 
                "Nitrogen_mg_l", "Nitrate_mg_l"]
cat_features = [col for col in X_train.columns if col not in num_features]

preprocessor = ColumnTransformer(transformers=[
    ('num', Pipeline([
        ('pw', PowerTransformer(method='yeo-johnson')),
        ('sc', StandardScaler())
    ]), num_features),
    ('cat', 'passthrough', cat_features)
])

X_train_proc = preprocessor.fit_transform(X_train)
X_val_proc   = preprocessor.transform(X_val)
X_test_proc  = preprocessor.transform(X_test)
```

```python
def create_sequences_fast(X_proc, y_data, areas_df, seq_length=5,stride=3):
    all_x, all_y = [], []
    areas_values = areas_df['Area'].values
    y_values = y_data.values if hasattr(y_data, 'values') else y_data
    unique_areas = areas_df['Area'].unique()
    pbar = tqdm(unique_areas, desc="Creating Augmented Sequences", unit="area")
    
    for area in pbar:
        mask = (areas_values == area)
        X_area = X_proc[mask]
        y_area = y_values[mask]
        
        if len(X_area) > seq_length:
            for start in range(0, len(X_area) - seq_length + 1, stride):
                all_x.append(X_area[start:start + seq_length])
                all_y.append(y_area[start + seq_length - 1])
    
    pbar.close()
    return np.array(all_x, dtype=np.float32), np.array(all_y, dtype=np.int64)
 
print("Starting sequence generation...")

start_time = time.time()

X_train_seq, y_train_seq = create_sequences_fast(
    X_train_proc, y_train, df_train_final, 
    seq_length=8,   
    stride=4       
)

X_val_seq, y_val_seq = create_sequences_fast(
    X_val_proc, y_val, df_val_final, 
    seq_length=8, 
    stride=1        
)

X_test_seq, y_test_seq = create_sequences_fast(
    X_test_proc, y_test, df_test_final, 
    seq_length=8, 
    stride=1       
)

print(f"Conversion completed in {time.time() - start_time:.2f} seconds")

class_weights_np = compute_class_weight(
    'balanced',
    classes=np.unique(y_train_seq),
    y=y_train_seq
)
class_weights = torch.tensor(class_weights_np, dtype=torch.float32).to(device)

print("\nFinal class weights:")

for cls_idx, w in enumerate(class_weights):
    print(f"{cls_idx} ({class_order[cls_idx]}): {w:.4f}")
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
class BidirectionalInputAttention(nn.Module):
 
    def __init__(self, input_size, hidden_size):
        super(BidirectionalInputAttention, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.W_f = nn.Linear(2*hidden_size, hidden_size, bias=False)
        self.U_f = nn.Linear(1, hidden_size, bias=False)
        self.v_f = nn.Parameter(torch.randn(hidden_size, 1))
        self.W_b = nn.Linear(2*hidden_size, hidden_size, bias=False)
        self.U_b = nn.Linear(1, hidden_size, bias=False)
        self.v_b = nn.Parameter(torch.randn(hidden_size, 1))
       
        nn.init.xavier_uniform_(self.v_f)
        nn.init.xavier_uniform_(self.v_b)
   
    def forward(self, X_t, h_prev_f, c_prev_f, h_prev_b, c_prev_b):
        batch_size = X_t.size(0)
        hc_f = torch.cat([h_prev_f, c_prev_f], dim=1)  
        W_output = self.W_f(hc_f)  
        energy_f_list = []
        for l in range(self.input_size):
            x_l = X_t[:, l:l+1]  
            U_output = self.U_f(x_l) 
            energy_l = torch.tanh(W_output + U_output)  
            energy_f_list.append(torch.matmul(energy_l, self.v_f))  
       
        energy_f = torch.cat(energy_f_list, dim=1) 
        alpha_f = F.softmax(energy_f, dim=1)  
        X_weighted_f = X_t * alpha_f  
       
         
        hc_b = torch.cat([h_prev_b, c_prev_b], dim=1)
        W_output_b = self.W_b(hc_b)  
       
        energy_b_list = []
        for l in range(self.input_size):
            x_l = X_t[:, l:l+1]
            U_output = self.U_b(x_l)
            energy_l = torch.tanh(W_output_b + U_output)
            energy_b_list.append(torch.matmul(energy_l, self.v_b))
       
        energy_b = torch.cat(energy_b_list, dim=1)
        alpha_b = F.softmax(energy_b, dim=1)
        X_weighted_b = X_t * alpha_b
       
        return X_weighted_f, X_weighted_b, alpha_f, alpha_b



class BidirectionalTemporalAttention(nn.Module):
    def __init__(self, encoder_hidden_size, decoder_hidden_size):

        super(BidirectionalTemporalAttention, self).__init__()
        self.W_o_f = nn.Linear(2*decoder_hidden_size, encoder_hidden_size, bias=False)

        self.U_o_f = nn.Linear(2*encoder_hidden_size, encoder_hidden_size, bias=False)

        self.v_o_f = nn.Parameter(torch.randn(encoder_hidden_size, 1))
        self.W_o_b = nn.Linear(2*decoder_hidden_size, encoder_hidden_size, bias=False)

        self.U_o_b = nn.Linear(2*encoder_hidden_size, encoder_hidden_size, bias=False)

        self.v_o_b = nn.Parameter(torch.randn(encoder_hidden_size, 1))
        nn.init.xavier_uniform_(self.v_o_f)
        nn.init.xavier_uniform_(self.v_o_b)

   
    def forward(self, encoder_outputs, h_decoder_f, h_decoder_b, s_decoder_f, s_decoder_b):

        ds_decoder_f = torch.cat([h_decoder_f, s_decoder_f], dim=1)
        W_f_out = self.W_o_f(ds_decoder_f) 
        U_f_out = self.U_o_f(encoder_outputs) 

        energy_f = torch.tanh(
            W_f_out.unsqueeze(1) + U_f_out 
        ) 
        beta_f = F.softmax(
            torch.matmul(energy_f, self.v_o_f).squeeze(-1), dim=1
        ) 
        context_f = torch.sum(
            beta_f.unsqueeze(-1) * encoder_outputs,
            dim=1
        ) 
        ds_decoder_b = torch.cat([h_decoder_b, s_decoder_b], dim=1)
        W_b_out = self.W_o_b(ds_decoder_b)
        U_b_out = self.U_o_b(encoder_outputs)
        energy_b = torch.tanh(
            W_b_out.unsqueeze(1) + U_b_out
        )
        beta_b = F.softmax(
            torch.matmul(energy_b, self.v_o_b).squeeze(-1), dim=1
        )
        context_b = torch.sum(
            beta_b.unsqueeze(-1) * encoder_outputs,
            dim=1
        )  
        return context_f, context_b, beta_f, beta_b



class VBAEDClassifier(nn.Module):

 

   

    def __init__(self, input_size, seq_length, num_classes,

                 encoder_hidden=64, decoder_hidden=64, dropout=0.3):

        super(VBAEDClassifier, self).__init__()

       

        self.input_size = input_size

        self.seq_length = seq_length

        self.num_classes = num_classes

        self.encoder_hidden = encoder_hidden

        self.decoder_hidden = decoder_hidden
        self.input_attention = BidirectionalInputAttention(

            input_size, encoder_hidden

        )
        self.lstm_encoder_f = nn.LSTMCell(
            input_size=input_size,
            hidden_size=encoder_hidden
        )  
        self.lstm_encoder_b = nn.LSTMCell(
            input_size=input_size,
            hidden_size=encoder_hidden
        )
        self.temporal_attention = BidirectionalTemporalAttention(
            encoder_hidden, decoder_hidden
        )
        self.w_decoder_f = nn.Linear(2*encoder_hidden, decoder_hidden, bias=True)  
        self.w_decoder_b = nn.Linear(2*encoder_hidden, decoder_hidden, bias=True)  
        self.lstm_decoder_f = nn.LSTMCell(
            input_size=decoder_hidden,
            hidden_size=decoder_hidden
        )
        self.lstm_decoder_b = nn.LSTMCell(
            input_size=decoder_hidden,
            hidden_size=decoder_hidden
        )
        final_dim = 2*decoder_hidden + 2*2*encoder_hidden
        self.fc1 = nn.Linear(final_dim, 256)
        self.bn1 = nn.BatchNorm1d(256)
        self.dropout1 = nn.Dropout(dropout)
        self.fc2 = nn.Linear(256, 128)
        self.bn2 = nn.BatchNorm1d(128)
        self.dropout2 = nn.Dropout(dropout)
        self.fc3 = nn.Linear(128, num_classes)
        self.relu = nn.ReLU()
    def forward(self, X):
        batch_size = X.size(0)
        seq_len = X.size(1)
        device = X.device
        h_encoder_f = torch.zeros(batch_size, self.encoder_hidden, device=device)
        c_encoder_f = torch.zeros(batch_size, self.encoder_hidden, device=device)
        h_encoder_b = torch.zeros(batch_size, self.encoder_hidden, device=device)
        c_encoder_b = torch.zeros(batch_size, self.encoder_hidden, device=device)
        encoder_outputs_f = []
        encoder_outputs_b = []

        for t in range(seq_len):

            x_t = X[:, t, :]
            x_weighted_f, x_weighted_b, alpha_f, alpha_b = self.input_attention(

                x_t, h_encoder_f, c_encoder_f, h_encoder_b, c_encoder_b

            )
            x_input_f = x_weighted_f
            x_input_b = x_weighted_b
            h_encoder_f, c_encoder_f = self.lstm_encoder_f(x_input_f, (h_encoder_f, c_encoder_f))
            encoder_outputs_f.append(h_encoder_f)
        for t in range(seq_len-1, -1, -1):
            x_t = X[:, t, :]
            x_weighted_f, x_weighted_b, alpha_f, alpha_b = self.input_attention(
                x_t, h_encoder_f, c_encoder_f, h_encoder_b, c_encoder_b

            )
            x_input_b = x_weighted_b
            h_encoder_b, c_encoder_b = self.lstm_encoder_b(x_input_b, (h_encoder_b, c_encoder_b))
            encoder_outputs_b.insert(0, h_encoder_b)
        encoder_outputs_f = torch.stack(encoder_outputs_f, dim=1)
        encoder_outputs_b = torch.stack(encoder_outputs_b, dim=1)
        encoder_outputs_full = torch.cat([encoder_outputs_f, encoder_outputs_b], dim=2)
        h_decoder_f = h_encoder_f
        s_decoder_f = c_encoder_f
        h_decoder_b = h_encoder_b
        s_decoder_b = c_encoder_b
        decoder_outputs_f = []
        decoder_outputs_b = []
        for t in range(seq_len):
            context_f, context_b, beta_f, beta_b = self.temporal_attention(
                encoder_outputs_full,
                h_decoder_f, h_decoder_b,
                s_decoder_f, s_decoder_b

            )
            decoder_input_f = self.w_decoder_f(context_f)   
            h_decoder_f, s_decoder_f = self.lstm_decoder_f(
                decoder_input_f,
                (h_decoder_f, s_decoder_f)
            )
            decoder_outputs_f.append(h_decoder_f)
        for t in range(seq_len-1, -1, -1):
            context_f, context_b, beta_f, beta_b = self.temporal_attention(
                encoder_outputs_full,
                h_decoder_f, h_decoder_b,
                s_decoder_f, s_decoder_b
            )
            decoder_input_b = self.w_decoder_b(context_b)   
            h_decoder_b, s_decoder_b = self.lstm_decoder_b(
                decoder_input_b,
                (h_decoder_b, s_decoder_b)
            )
            decoder_outputs_b.insert(0, h_decoder_b)
        d_f_T = decoder_outputs_f[-1]
        d_b_1 = decoder_outputs_b[0]
        features = torch.cat([
            d_f_T,
            context_f,
            d_b_1,
            context_b
        ], dim=1)
        x = self.fc1(features)
        x = self.bn1(x)
        x = self.relu(x)
        x = self.dropout1(x)
        x = self.fc2(x)
        x = self.bn2(x)
        x = self.relu(x)
        x = self.dropout2(x)
        logits = self.fc3(x)
        return logits

 
 

     
```

```python

input_size = X_train_t.shape[-1]   
model = VBAEDClassifier(
    input_size=input_size,
    seq_length=8,            
    num_classes=4,           
    encoder_hidden=128,      
    decoder_hidden=128,
    dropout=0.4             
).to(device)

 
class_weights = torch.tensor([1.0, 1.2, 1.5, 2.0], device=device)   
criterion = nn.CrossEntropyLoss(weight=class_weights)
optimizer = torch.optim.AdamW(model.parameters(), lr=0.0005, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5)
```

```python
  

plt.rcParams['font.weight'] = 'bold'
plt.rcParams['font.size'] = 11
sns.set_style("whitegrid")

 
optimizer = optim.AdamW(model.parameters(), lr=0.0005, weight_decay=1e-5)
criterion = nn.CrossEntropyLoss(weight=class_weights.to(device))  
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='max', factor=0.5, patience=5, verbose=True
)

num_epochs = 20   
patience = 12      
min_delta = 0.5   
 
train_losses, train_accs, train_bal_accs = [], [], []
val_losses, val_accs, val_bal_accs = [], [], []

best_val_bal_acc = 0.0
patience_counter = 0

print("=" * 100)
print(f"{'Epoch':<8} | {'Train Loss':<12} | {'Train Acc':<12} | {'Train Bal.Acc':<14} | "
      f"{'Val Loss':<12} | {'Val Acc':<12} | {'Val Bal.Acc':<14} | {'LR':<10}")
print("-" * 100)

for epoch in range(num_epochs):
    model.train()
    train_loss = 0.0
    train_correct = 0
    train_total = 0
    train_preds = []
    train_targets = []

     
    loop = tqdm(train_loader, leave=False)
    for batch_x, batch_y in loop:
        batch_x, batch_y = batch_x.to(device), batch_y.to(device)

        optimizer.zero_grad()
        logits = model(batch_x)  
        loss = criterion(logits, batch_y)

        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)   
        optimizer.step()

        train_loss += loss.item() * batch_y.size(0)
        preds = torch.argmax(logits, dim=1)
        train_correct += (preds == batch_y).sum().item()
        train_total += batch_y.size(0)

        train_preds.extend(preds.cpu().numpy())
        train_targets.extend(batch_y.cpu().numpy())

    
    train_loss /= train_total
    train_acc = 100 * train_correct / train_total
    train_bal_acc = 100 * balanced_accuracy_score(train_targets, train_preds)

    train_losses.append(train_loss)
    train_accs.append(train_acc)
    train_bal_accs.append(train_bal_acc)

     
    model.eval()
    val_loss = 0.0
    val_correct = 0
    val_total = 0
    val_preds = []
    val_targets = []

    with torch.no_grad():
        for batch_x, batch_y in val_loader:
            batch_x, batch_y = batch_x.to(device), batch_y.to(device)
            logits = model(batch_x)
            loss = criterion(logits, batch_y)

            val_loss += loss.item() * batch_y.size(0)
            preds = torch.argmax(logits, dim=1)
            val_correct += (preds == batch_y).sum().item()
            val_total += batch_y.size(0)

            val_preds.extend(preds.cpu().numpy())
            val_targets.extend(batch_y.cpu().numpy())

    val_loss /= val_total
    val_acc = 100 * val_correct / val_total
    val_bal_acc = 100 * balanced_accuracy_score(val_targets, val_preds)

    val_losses.append(val_loss)
    val_accs.append(val_acc)
    val_bal_accs.append(val_bal_acc)

    current_lr = optimizer.param_groups[0]['lr']

    print(f"{epoch+1:<8} | {train_loss:<12.4f} | {train_acc:<12.2f} | {train_bal_acc:<14.2f} | "
          f"{val_loss:<12.4f} | {val_acc:<12.2f} | {val_bal_acc:<14.2f} | {current_lr:<10.6f}")

     
    scheduler.step(val_bal_acc)

     
    if val_bal_acc > best_val_bal_acc + min_delta:
        best_val_bal_acc = val_bal_acc
        patience_counter = 0
        torch.save(model.state_dict(), 'best_vbaed_classifier.pth')
        print(f"   Saved best model! (Val Bal.Acc: {val_bal_acc:.2f}%)")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print(f"\nEarly Stopping at epoch {epoch+1} (no improvement for {patience} epochs)")
            break

print("=" * 100)
print(f"Best Validation Balanced Accuracy: {best_val_bal_acc:.2f}%")
print("Training finished!")
```
